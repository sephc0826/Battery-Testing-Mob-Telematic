```python
link=("/content/drive/MyDrive/Product/Battery Drainage Investigation /BattResponses2.0.csv")
```

```python
df = pd.read_csv(link, index_col='Timestamp',
                 parse_dates=True, dayfirst=True)
```

```python
df.describe()
```

```python
ip = df.Mob.str.contains(pat = 'ip|Ip|iP|IO|IP')
iphone = df.loc[ip]
iphone.head(5)
```

```python
not_iphone = df.Mob.str.contains(pat = 'ip|Ip|iP|IO|IP|test') == False
android = df.loc[not_iphone]
android.head(5)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=4, cols=1)

fig.add_trace(
    go.Scatter(
        name ="Nagase",
        x=iphone.Mob,
        y=iphone.Nagase,
        mode="markers"),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        name ="MSIG",
        x=iphone.Mob,
        y=iphone.MSIG,
        mode="markers"),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(
        name ="SMAT",
        x=iphone.Mob,
        y=iphone.SMAT,
        mode="markers"),
    row=3, col=1
)

fig.add_trace(
    go.Scatter(
        name ="Tracker",
        x=iphone.Mob,
        y=iphone.Tracker,
        mode="markers"),
    row=4, col=1
)

fig.update_layout(height=1000, width=1000, title_text="Battery Usage - IOS")

fig.show()
```
